## Functions

### adjusted_r2

Docstring

 """
    Compute adjusted R-squared for a model.

    Parameters
    ----------
    y_true : array-like - True, raw data
    y_pred : array-like - Fit predictions
    n_params : int - Number of parameters in fit

    Returns
    -------
    r2_adj : float - Adjusted R-squared.
    """

This function calculates the adjusted R-squared value for a given fit and data, using the following equation:

**$$R^2_{\text{adj}} = 1 - \left( \frac{(1 - R^2)(n - 1)}{n - p - 1} \right)$$**

**Where**
- $n$ = number of data points
- $p$ = number of parameters in fit

```python
X = np.arange(360) # Range of x value for model
params = params_decomped[0,0] # parameters from fitting for pixel 0,0
data = simulated_dataset[0][0,0] # Data from pixel 0,0

assess.adjusted_r2(data,fit.periodic(X,*params),5)
```

### signal_leakage

Docstring

"""
    Total raw variance captured in the residuals as fraction.

    Parameters
    ----------
    raw : array-like - True, raw data
    denoised : array-like - Denosied data

    Returns
    -------
    float - Signal leakage value.
"""

Calculates the fraction of the raw signal variance contained in the residual, using:

$$
\text{signal\_leakage} =
\frac{\mathrm{Var}(x_{\text{raw}} - x_{\text{denoised}})}
{\mathrm{Var}(x_{\text{raw}})}
$$

#### Where
- $x_{\text{raw}}$ = original signal
- $x_{\text{denoised}}$ = denoised signal
- $\mathrm{Var}(\cdot)$ = variance operator

```python
raw_data = sd[1]
cleaned_data = sd[0]

assess.signal_leakage(raw_data,cleaned_data)
```
