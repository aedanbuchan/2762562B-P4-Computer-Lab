# Asess Demo

This notebook demonstrates the `PhysComp.assess` module, which assesses the decomposition and fitting

### Functions covered:
- `assess.adjusted_r2` — generates adjusted $R^2$ values for a given fit and data
- `assess.signal_leakage` — Calculates raw varience in residuals

```python
### Needed imports for functions
import numpy as np

### Importing simulation and fitting function for example data
import PhysComp.sim as sim
import PhysComp.uml as uml
import PhysComp.fitting as fit

### Importing the visualise function
import PhysComp.assess as assess
```

#### Generating simulated data, decomposing and fitting for example data to assess. See respective demo notebooks for more information.

```python
simulated_dataset = sim.sim_dataset(30,30,360,10,10, filesave= True,save_path="Test.npy", seed=100) # Simulating data, see Sim notebook
sd = uml.decomp_data("Test.npy","SVD")
params_decomped, errors_decomped = fit.periodic_fit_whole(sd[0], [1,1,1,1,1])
params_raw, errors_raw = fit.periodic_fit_whole(sd[1], [1,1,1,1,1]) 
```

### adjusted_r2 : Calculates the adjusted $R^2$ value for given fit and data
#### $$R^2_{\text{adj}} = 1 - \left( \frac{(1 - R^2)(n - 1)}{n - p - 1} \right)$$

#### Where
- $n$ = number of data points
- $p$ = number of parameters in fit

```python
X = np.arange(360) # Range of x value for model
params = params_decomped[0,0] # parameters from fitting for pixel 0,0
data = simulated_dataset[0][0,0] # Data from pixel 0,0

assess.adjusted_r2(data,fit.periodic(X,*params),5)
```

### signal_leakage
Fraction of the raw signal variance contained in the residual.

$$
\text{signal\_leakage} =
\frac{\mathrm{Var}(x_{\text{raw}} - x_{\text{denoised}})}
{\mathrm{Var}(x_{\text{raw}})}
$$

#### Where
- $x_{\text{raw}}$ = original signal
- $x_{\text{denoised}}$ = denoised signal
- $\mathrm{Var}(\cdot)$ = variance operator

```python
raw_data = sd[1]
cleaned_data = sd[0]

assess.signal_leakage(raw_data,cleaned_data)
```
